### chi-square test

#### A chi-square statistic is a measure of the difference between the observed and expected frequencies of the outcomes of a set of events or variables.
#### Chi-square is useful for analyzing such differences in categorical variables, especially those nominal in nature.

#### H0 - The two variable are independent means that there is no relationship between two categorical variables
#### HA - There is relationship between two categorical variable they are dependent 

```python
#check chisquare between job and response
df_table = pd.crosstab(df1_sample['job'],df1_sample['response'])
print(df_table)
```

```python
#observed values
observed_values = df_table.values
print('observed values:- \n',observed_values)
```

```python
val = stats.chi2_contingency(df_table)
```

```python
val
```

```python
#excepted values
expected_values = val[3]
```

```python
expected_values 
```

```python
no_of_rows = len(df_table.iloc[0:,0])
no_of_columns = len(df_table.iloc[0,0:])
ddof = (no_of_rows-1)*(no_of_columns-1)
print('Degree of Freedom:-',ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square = sum([(o-e)**2./e for o,e in zip(observed_values,expected_values)])
chi_square_statistic = chi_square[0]+chi_square[1]
```

```python
print('chi-square statistic:-',chi_square_statistic)
```

```python
critical_value = chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

```python
#p-value
p_value = 1 - chi2.cdf(x = chi_square_statistic,df=ddof)
print('p-value: ',p_value)
print('significance level: ',alpha)
print('Degree of freedom: ',ddof)
```

```python
p_value
```

```python
if chi_square_statistic>=critical_value:
    print('Reject H0,There is a relationship between 2 categorical variables')
else:
    print('Fail to reject H0, There is no relationship between 2 categorial variables')
    
if p_value <= alpha:
    print('Reject H0,There is a relationship between 2 categorical variables')
else:
    print('Fail to reject H0, There is no relationship between 2 categorial variables')    
    
```

### Find the relation of salary and age column using statistical tests and draw the observations from it

### T test

#### A t-test is a statistical test that compares the means of two samples. It is used in hypothesis testing, with a null hypothesis that the difference in group means is zero and an alternate hypothesis that the difference in group means is different from zero

#### In this finding the relation of salary and age column mean so we use two sample t test

```python
_,p_value = stats.ttest_ind(a = df1['age'],b=df1['salary'],equal_var=False)
```

```python
if p_value < 0.05:
    print('Reject the null hypothesis,there is difference between mean of this two variables ')
else:
    print('Fail to Reject the null hypothesis,there is no difference between mean of this two variables ')
```

#### By using this test,find that there is difference between mean of this two variables that means they are statistical significant means that there is relationship between this two variables
